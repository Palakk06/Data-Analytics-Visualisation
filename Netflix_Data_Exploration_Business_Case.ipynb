```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Basic Analysis
# 1. Un-nesting the columns
# a. Un-nest the columns those have cells with multiple comma separated values by creating multiple rows

df = pd.read_csv('/content/sample_data/netflix_titles.csv') # Read the dataset
df
```

```python
#No. of row and columns in a dataframe
df.shape
```

```python
# 2. Handling null values
# a. For categorical variables with null values, update those rows as
# unknown_column_name.

#checking dataframe and getting information about the typre of data
df.info()
```

```python
#Deleting Null values duration,rating and date_added
df.dropna(subset = ["duration"], inplace = True)
df.dropna(subset = ["rating"], inplace = True)
df.dropna(subset = ["date_added"], inplace = True)
```

```python
#Handling Null values for countries, director and cast
df["country"] = df["country"].fillna("Unknown Country")
df["cast"] = df.cast.fillna("Unknown Cast")
df["director"] = df.director.fillna("Unknown Director")
```

```python
#Checking data info after handling null values
df.info()
```

```python
# Find the counts of each categorical variable both using graphical and nongraphical analysis.
# a. For Non-graphical Analysis:
# b. For graphical analysis:

#performing the statistical analysis on the numerical column
df.describe()
```

```python
df.describe(include = ["object"])
```

```python
#Checking for duplicates
df.duplicated().sum()
```

```python
#Separating the directors based on commas so I can have exact number of count of directors
drct = df.assign(director = df['director'].str.split(' , ')).explode('director', ignore_index = True).copy()
Pop_drct = drct["director"].value_counts()[:10]
Pop_drct
```

```python
#Plotting the graph for Most popular directors and directors
drct = sns.barplot(x = Pop_drct.index, y = Pop_drct.values)
drct.bar_label(drct.containers[0])
plt.title("Most Popular directors")
plt.xlabel("Directors")
plt.xticks(rotation = 90)
plt.show()
```

```python
# 2. Comparison of tv shows vs. movies.
# a. Find the number of movies produced in each country and pick the top 10 countries.
# b. Find the number of Tv-Shows produced in each country and pick the top 10 countries.

# distribution of data in terms of movies and Tv Shows
df["type"].value_counts()
```

```python
#Separating the countries to retrieve the exact count of countries based on the content produced
cnt = df.assign(country=df['country'].str.split(', ')).explode('country', ignore_index = True).copy()
```

```python
#top 10 countries based on the content produced
countries = cnt.country.value_counts()[:10]
countries
```

```python
#Plotting Most popular countries for content produced on graph
labs = sns.barplot(x = countries.index, y = countries.values)
plt.xticks(rotation = 90)
plt.xlabel("Countries")
plt.title("Most popular countries for Content")
labs.bar_label(labs.containers[0])
plt.show()
```

```python
#Analysis of release year
release_yr = df.release_year.value_counts().sort_values(ascending = False)
release_yr
```

```python
#Graphical Plotting of release years (top 10)
release_yr = sns.barplot(x = release_yr[:10].index, y = release_yr[:10].values)
plt.title("Graphical Analysis of Content Release distribution")
plt.xlabel("Release Year")
plt.xticks(rotation = 90)
release_yr.bar_label(release_yr.containers[0])
plt.show()
```

```python
#Distribution of content by rating provided
ratings = df.rating.value_counts()
ratings
```

```python
#Graphical representation of content vs Ratings
plt.title("Content distribution Vs Rating")
ratings = sns.barplot(x = ratings.index, y = ratings.values)
ratings.bar_label(ratings.containers[0])
plt.xticks(rotation = 90)
plt.xlabel("Rating")
plt.show()
```

```python
#Separating the dataframe based on the type of content
movies = df.loc[df["type"] == "Movie"].reset_index(drop = True)
tv_show = df.loc[df["type"] == "TV Show"].reset_index(drop = True)
```

```python
#separating the countries based on commas so I can have exact number of count of countries for movies
mv = movies.assign(country = movies['country'].str.split(', ')).explode('country', ignore_index = True).copy()
mv = mv.reset_index(drop = True).copy()
mv
```

```python
#Top 10 countries for producing the movies in the country
pop_mv_country = mv.groupby("country")["show_id"].agg("count").sort_values(ascending = False)[:10]
pop_mv_country
```

```python
#plotting the most popular countries vs the count of content produced on a graph
mv_cnt = sns.barplot(x = pop_mv_country.index, y = pop_mv_country.values)
plt.xticks(rotation = 90)
mv_cnt.bar_label(mv_cnt.containers[0])
plt.title("Popular countries for movies")
plt.show()
```

```python
# 3. What is the best time to launch a TV show?
# a. Find which is the best week to release the Tv-show or the movie. Do the analysis separately for Tv-shows and Movies
# b. Find which is the best month to release the Tv-show or the movie. Do the analysis separately for Tv-shows and Movies

#separating the countries using comma as a delimiter
tvs = tv_show.assign(country = tv_show['country'].str.split(', ')).explode('country', ignore_index = True).copy()
tvs = tvs.reset_index(drop = True).copy()
```

```python
#finding the top 10 countries as per the release of content in the specific country
top_cont_tvs = tvs.groupby("country")["show_id"].agg("count").sort_values(ascending = False)[:10]
top_cont_tvs
```

```python
#plotting top 10 countries vs count of tv shows produced on a graph
plt.title("Popular countries for Tv Shows")
pop_cont_tvs = sns.barplot(x = top_cont_tvs.index, y = top_cont_tvs.values)
plt.xticks(rotation = 90)
pop_cont_tvs.bar_label(pop_cont_tvs.containers[0])
plt.show()
```

```python
# Converting date_added column to datetime to do analysis for most suitable month for releasing a movie
movies["date_added"] = pd.to_datetime(movies["date_added"])
```

```python
#Extracting the week from the date on which the content was uploaded on netflix
movies["week"] = (movies["date_added"].dt.strftime('%U').astype(int) + 1)
movies.head()
```

```python
#Most popular weeks for releasing a movie
movies["week"].value_counts()[:10]
```

```python
#Most popular months to release a movie
movies["Month"] = movies.date_added.dt.month
movies["Month"].value_counts()[:10]
```

```python
#converting date added for tv shows to do time analysis

# Convert "date_added" column to datetime with mixed format
tv_show["date_added"] = pd.to_datetime(tv_show["date_added"], format='mixed')

# Extract week number starting from 1
tv_show["week"] = (tv_show["date_added"].dt.strftime('%U').astype(int) + 1)
tv_show
```

```python
#Most popular weeks to release a tv show
tv_show.week.value_counts()[:10]
```

```python
#Most Popular Month to release a tv show
tv_show["Month"] = tv_show.date_added.dt.month
tv_show.Month.value_counts()[:10]
```

```python
# Insight: After Analysis, it turns out that December is most popular month for releasing a tv show followed by July and September
```

```python
# 4. Analysis of actors/directors of different types of shows/movies.
# a. Identify the top 10 directors who have appeared in most movies or TV shows.
# b. Identify the top 10 directors who have appeared in most movies or TV shows.

#Top_10 actors for movies or tv_shows
cast = df.assign(cast = df["cast"].str.split(", ")).explode("cast", ignore_index = True).copy()
```

```python
#Value count for top 10 actors
top_10_actors = cast["cast"].value_counts()[:10]
top_10_actors
```

```python
#plotting the top 10 actors on a graph
cast2 = sns.barplot(x = top_10_actors.index, y = top_10_actors.values)
plt.xticks(rotation = 90)
plt.title("Top 10 Actors")
plt.xlabel("Actors")
plt.ylabel("Movie Counts")
cast2.bar_label(cast2.containers[0])
plt.show()
```

```python
#Top_10 directors for movies and tv_shows
top_drct= df.assign(director = df.director.str.split(", ")).explode("director", ignore_index = True).copy()
top_10_drct = top_drct["director"].value_counts()[:10]
top_10_drct
```

```python
#plotting top 10 directors on a graph for graphical analysis
pop_drct = sns.barplot(x = top_10_drct.index, y = top_10_drct.values)
plt.title("Top 10 directors")
plt.xlabel("Directors")
plt.ylabel("Count of movies/tv shows")
plt.xticks(rotation = 90)
pop_drct.bar_label(pop_drct.containers[0])
plt.show()
```

```python
# Insight: the most popular actors are Anupanm Kher, Shah Rukh Khan, Julie Tejwani and so on
```

```python
# 5. Which genre movies are more popular or produced more

#most popular movie genre
top_movie = movies.assign(listed_in = movies["listed_in"].str.split(", ")).explode("listed_in", ignore_index = True)
top_10_mv_genre = top_movie["listed_in"].value_counts()[:10]
top_10_mv_genre
```

```python
#plotting the top 10 gernres on a graph
top_movs_genre = sns.barplot(x = top_10_mv_genre.index, y = top_10_mv_genre.values)
top_movs_genre.bar_label(top_movs_genre.containers[0])
plt.xlabel("Genres")
plt.title("Top 10 Movie Genres")
plt.xticks(rotation = 90)
plt.show()
```

```python
# Insight: seems that the Rajiv Chilaka is the most popular director, followed by Jan Suter, RaÃºl Campos and so on
```

```python
# 6. Find After how many days the movie will be added to Netflix after the release of the movie (you can consider the recent past data)

from wordcloud import WordCloud
import matplotlib.pyplot as plt

gens = []
for i in top_movie["listed_in"]:
    gens.append(i)

gens1 = ",".join(gens)
wordcloud = WordCloud(width=1200, height=700, background_color="black").generate(gens1)
plt.imshow(wordcloud)
plt.title("Most popular movie genres")
plt.figure(figsize=(12, 5))
plt.axis("off")
plt.show()
```

```python
#After how many days the movie will be added to Netflix after the release of the movie
movies['days_to_added'] = (pd.to_datetime(movies['date_added']) - pd.to_datetime(movies['release_year'], format='%Y')).dt.days
movies["days_to_added"].mode()
```

```python
#Insight: After analysis, it seems that it takes 334 days for a movie to release on netflix after it has been released in theaters
```
